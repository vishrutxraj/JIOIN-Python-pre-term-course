```python
class Aircraft:
  def __init__(self,reg,range_NM,power,make,model,life):
    self.reg = reg
    self.range_NM = range_NM
    self.power = power
    self.make = make
    self.model = model
    self.life = life

  def display(self):
   print(self.reg,self.range_NM,self.power,self.make,self.model,self.life)
```

```python
a1 = Aircraft("VT-RTN", 3300, "27000N", "Airbus", "A320", "20 years")
a2 = Aircraft("VT-ALL", 9500, "88000N", "Boeing", "777", "19 years")
a3 = Aircraft("VT-TSD", 12000, "89000N", "Boeing", "787", "42 years")
```
