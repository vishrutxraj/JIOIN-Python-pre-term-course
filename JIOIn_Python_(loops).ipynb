#FizzBuzz Problem

Print a sequence from 1 - 100,
But if you see a multiple of 3, print fizz,
And if you see a multiple of 5, print buzz,
If you see a multiple of 3 and 5, print fizzbuzz.

```python
for i in range (1,101) :
  if i % 15 == 0 :
    print("FizzBuzz")
  elif i % 3 == 0 :
    print("Fizz")
  elif i % 5 == 0:
    print("Buzz")
  else:print(i)
```
