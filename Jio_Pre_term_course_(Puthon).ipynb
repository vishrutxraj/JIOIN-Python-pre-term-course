```python

user_name = input("Please enter your name: ")
print(f"\nHello {user_name}! Welcome to Python Cafe ☕\n")


food_choice = input("Please type V for Veg or N for Non-Veg: ").upper()

while food_choice not in ["V", "N"]:
    print("Invalid input! Please enter V or N.")
    food_choice = input("Please type V for Veg or N for Non-Veg: ").upper()


add_spices = input("Press Y to add spices or N for no spices: ").upper()

while add_spices not in ["Y", "N"]:
    print("Invalid input! Please enter Y or N.")
    add_spices = input("Press Y to add spices or N for no spices: ").upper()

# Cafe menu
menu = {
    "Coffee": {"type": "V", "spices": "N"},
    "Misal": {"type": "V", "spices": "Y"},
    "Ice cream": {"type": "N", "spices": "N"},
    "Burger": {"type": "N", "spices": "Y"}
}

print("\nAvailable options for you:\n")


found = False

for item, details in menu.items():
    if details["type"] == food_choice and details["spices"] == add_spices:
        print(item)
        found = True

if not found:
    print("Sorry, no items match your preference.")

print("\nThank you for visiting Puthon Cafe!")
```

Dictionary Practice

Sample

```python
student = {
    "name": "Rahul",
    "age": 21,
    "course": "AIML",
    "marks": 85
}

print(student)
```

Print Keys

```python
print(student.keys())
```

Print Values

```python
print(student.values())
```

Accessing

```python
print(student["name"])
print(student.get("age"))
```

Adding Key and Value

```python
student["city"] = "Delhi"
print(student)
```

Updating value


```python
student["marks"] = 90
print(student)
```

Removing

```python
student.pop("age")      # remove specific key
print(student)

student.popitem()       # removes last item
print(student)
```

Check if Key Exists

```python
if "name" in student:
    print("Key exists")
```
