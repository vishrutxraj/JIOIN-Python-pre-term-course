NUMPY


```python
import numpy as np
a = np.array([3,430,69,82])
b = np.array([34,56,78,90])


add = np.add(a,b)
sub = np.subtract(a,b)
multiply = np.multiply(a,b)
divide = np.divide(a,b)

print(add)
print(sub)
print(multiply)
print(divide)
```
